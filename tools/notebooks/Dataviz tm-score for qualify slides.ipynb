```python
%matplotlib inline
from sklearn.manifold import TSNE
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ggplot import *
# import pyrosetta
import os
import csv
import sys

print(sys.version)

sns.set(style="whitegrid")
```

```python
data = pd.read_csv('tm-score.csv')
data
```

```python
sns.set(rc={'figure.figsize':(16, 10)})
plot = sns.barplot(x='protein', y='best', hue='method', data=data)
plot.set_title('Best TM-Score grouped by protein', fontsize=40)
plot.set_ylabel('TM-Score (Higher is better)', fontsize=20)
plot.set_xlabel('Protein', fontsize=20)
plot.tick_params(labelsize=25)

plt.setp(plot.get_legend().get_texts(), fontsize='22')
plt.setp(plot.get_legend().get_title(), fontsize='32')

plot.get_legend().set_title('Methods')

plot.get_figure().savefig('tmscore_best_energy.png')

plot
```

```python
plot = sns.barplot(x='protein', y='mean', hue='method', yerr=np.asarray(data['stddev']), data=data)
plot.set_title('Average TM-Score grouped by protein', fontsize=40)
plot.set_ylabel('TM-Score (Higher is better)', fontsize=20)
plot.set_xlabel('Protein', fontsize=20)
plot.tick_params(labelsize=25)

plt.setp(plot.get_legend().get_texts(), fontsize='22')
plt.setp(plot.get_legend().get_title(), fontsize='32')

plot.get_legend().set_title('Methods')

plot.get_figure().savefig('tmscore_mean_energy.png')

plot
```
